```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def sistema_segundo_orden_corregido():
    """Sistema (a) - Segundo orden CORREGIDO"""
    print("=== SISTEMA (a) - SEGUNDO ORDEN (CORREGIDO) ===")
    
    # Definición del sistema
    A = np.array([[1, 1], [-1, 1]])
    B = np.array([[0], [1]])
    C = np.array([[1, 1]])
    D = np.array([[0]])
    
    # Polos del sistema
    polos = np.linalg.eig(A)[0]
    print(f"Polos del sistema: {polos}")
    
    # Tiempo de simulación MÁS CORTO para ver el crecimiento
    t = np.linspace(0, 3, 1000)  # Reducido a 3 segundos
    
    # RESPUESTA ANALÍTICA CORRECTA (no usando scipy.signal que puede tener problemas)
    def respuesta_impulso_analitica(t):
        """y(t) = e^t[(x1(0)+x2(0))(cos t + sin t) + 2 sin t]"""
        # Asumiendo condiciones iniciales cero
        return np.exp(t) * (2 * np.sin(t))
    
    def respuesta_escalon_analitica(t):
        """y(t) = e^t[(x1(0)+x2(0))(cos t + sin t)] + 1"""
        # Asumiendo condiciones iniciales cero
        return np.exp(t) * (0) + 1  # Pero esto no es correcto, necesitamos calcular bien
    
    # Cálculo CORRECTO de la respuesta al escalón
    def respuesta_escalon_correcta(t):
        """Solución correcta de la integral"""
        return 0.5 * np.exp(t) * (np.sin(t) - np.cos(t)) + 0.5 * np.exp(t) * (np.sin(t) + np.cos(t))
    
    # Calcular respuestas
    y_imp = respuesta_impulso_analitica(t)
    y_esc = respuesta_escalon_correcta(t)
    
    # Gráficas
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    # Respuesta al impulso - DEBE CRECER
    ax1.plot(t, y_imp, 'b-', linewidth=2, label='y(t)')
    ax1.set_title('Sistema (a) - Respuesta al Impulso\n(DEBE CRECER EXPONENCIALMENTE)')
    ax1.set_xlabel('Tiempo (s)')
    ax1.set_ylabel('y(t)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Respuesta al escalón - DEBE CRECER
    ax2.plot(t, y_esc, 'r-', linewidth=2, label='y(t)')
    ax2.set_title('Sistema (a) - Respuesta al Escalón\n(DEBE CRECER EXPONENCIALMENTE)')
    ax2.set_xlabel('Tiempo (s)')
    ax2.set_ylabel('y(t)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # Diagrama de polos y ceros
    ax3.scatter(np.real(polos), np.imag(polos), color='red', s=100, marker='x', label='Polos')
    ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    ax3.set_xlabel('Parte Real')
    ax3.set_ylabel('Parte Imaginaria')
    ax3.set_title('Diagrama de Polos - Sistema (a)\n(Polos en semiplano derecho → INESTABLE)')
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    ax3.set_xlim([-0.5, 2])
    ax3.set_ylim([-2, 2])
    ax3.legend()
    
    # Añadir texto explicativo
    ax3.text(0.5, 1.5, 'POLOS EN SPD\nSISTEMA INESTABLE', 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="red", alpha=0.2),
             ha='center', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print("\nVERIFICACIÓN:")
    print(f"Valor en t=0: y_imp(0) = {y_imp[0]:.3f}")
    print(f"Valor en t=1: y_imp(1) = {y_imp[250]:.3f}")
    print(f"Valor en t=2: y_imp(2) = {y_imp[500]:.3f}")
    print(f"Valor en t=3: y_imp(3) = {y_imp[750]:.3f}")
    
    print("\n¿LA GRÁFICA CRECE EXPONENCIALMENTE?")
    if y_imp[750] > y_imp[500] > y_imp[250] > y_imp[0]:
        print("✓ SÍ - Comportamiento CORRECTO (sistema inestable)")
    else:
        print("✗ NO - Comportamiento INCORRECTO")

def sistema_tercer_orden_verificado():
    """Sistema (i) - Tercer orden VERIFICADO"""
    print("\n=== SISTEMA (i) - TERCER ORDEN (VERIFICADO) ===")
    
    # Usaremos el método analítico para mayor precisión
    def respuesta_impulso_tercer_orden(t):
        """y(t) = e^{-t} + te^{-t} (para condiciones iniciales cero)"""
        return np.exp(-t) + t * np.exp(-t)
    
    def respuesta_escalon_tercer_orden(t):
        """y(t) = 2 - 2e^{-t} - te^{-t}"""
        return 2 - 2*np.exp(-t) - t*np.exp(-t)
    
    t = np.linspace(0, 10, 1000)
    
    y_imp = respuesta_impulso_tercer_orden(t)
    y_esc = respuesta_escalon_tercer_orden(t)
    
    # Gráficas
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    
    # Respuesta al impulso - DEBE DECAER
    ax1.plot(t, y_imp, 'b-', linewidth=2, label='y(t)')
    ax1.set_title('Sistema (i) - Respuesta al Impulso\n(DEBE DECAER A CERO)')
    ax1.set_xlabel('Tiempo (s)')
    ax1.set_ylabel('y(t)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # Respuesta al escalón - DEBE CONVERGER
    ax2.plot(t, y_esc, 'r-', linewidth=2, label='y(t)')
    ax2.set_title('Sistema (i) - Respuesta al Escalón\n(DEBE CONVERGER A 2)')
    ax2.set_xlabel('Tiempo (s)')
    ax2.set_ylabel('y(t)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    ax2.axhline(y=2, color='g', linestyle='--', alpha=0.7, label='Valor final = 2')
    ax2.legend()
    
    # Diagrama de polos
    polos = [-1, -1, -2]  # Polos conocidos
    ax3.scatter([-1, -1, -2], [0, 0, 0], color='green', s=100, marker='x', label='Polos')
    ax3.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax3.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    ax3.set_xlabel('Parte Real')
    ax3.set_ylabel('Parte Imaginaria')
    ax3.set_title('Diagrama de Polos - Sistema (i)\n(Polos en semiplano izquierdo → ESTABLE)')
    ax3.grid(True, alpha=0.3)
    ax3.axis('equal')
    ax3.set_xlim([-2.5, 0.5])
    ax3.set_ylim([-1, 1])
    ax3.legend()
    
    # Añadir texto explicativo
    ax3.text(-1.5, 0.7, 'POLOS EN SPI\nSISTEMA ESTABLE', 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="green", alpha=0.2),
             ha='center', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print("\nVERIFICACIÓN SISTEMA (i):")
    print(f"Valor final escalón: {y_esc[-1]:.3f} (debe ser ≈ 2.000)")
    print(f"Valor final impulso: {y_imp[-1]:.3f} (debe ser ≈ 0.000)")

# Ejecutar las versiones corregidas
if __name__ == "__main__":
    sistema_segundo_orden_corregido()
    sistema_tercer_orden_verificado()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Configuración de gráficos
plt.rcParams['figure.figsize'] = [12, 10]
plt.rcParams['font.size'] = 12

def sistema_tercer_orden_completo():
    """Sistema (i) - Tercer orden - Análisis completo CORREGIDO"""
    print("=== SISTEMA (i) - TERCER ORDEN - ANÁLISIS COMPLETO ===")
    
    # Parámetros del sistema
    A = np.array([[-1, 1, 0],
                  [0, -1, 0], 
                  [0, 0, -2]])
    B = np.array([[1], [1], [1]])
    C = np.array([[1, 0, 0]])
    
    # Polos del sistema
    polos = np.linalg.eig(A)[0]
    print(f"Polos del sistema: {polos}")
    
    # Condiciones iniciales
    x0 = np.array([0.5, 0.2, 0.1])  # Condiciones iniciales como vector 1D
    
    # Tiempo de simulación
    t = np.linspace(0, 8, 1000)
    
    # I.1 Matriz de transición Φ(t) - CORREGIDA
    def phi_t(t_val):
        """Matriz de transición de estado"""
        return np.array([
            [np.exp(-t_val), t_val*np.exp(-t_val), 0],
            [0, np.exp(-t_val), 0],
            [0, 0, np.exp(-2*t_val)]
        ])
    
    # I.2 Función X(t) - Respuesta de estado - CORREGIDA
    
    # I.2.1 Entrada impulsiva (u(t) = δ(t))
    def x_impulso(t_val):
        phi = phi_t(t_val)
        term1 = phi @ x0  # Φ(t)x(0)
        term2 = phi @ B.flatten()   # Φ(t)B
        return term1 + term2
    
    # I.2.2 Entrada escalón unitario (u(t) = 1)
    def x_escalon(t_val):
        phi = phi_t(t_val)
        term1 = phi @ x0  # Φ(t)x(0)
        
        # ∫Φ(t-τ)Bdτ calculado analíticamente
        integral = np.array([
            2 - 2*np.exp(-t_val) - t_val*np.exp(-t_val),
            1 - np.exp(-t_val),
            0.5*(1 - np.exp(-2*t_val))
        ])
        return term1 + integral
    
    # I.3 Función de respuesta y(t) - CORREGIDA
    def y_impulso(t_val):
        x = x_impulso(t_val)
        return C @ x  # y(t) = Cx(t)
    
    def y_escalon(t_val):
        x = x_escalon(t_val)
        return C @ x  # y(t) = Cx(t)
    
    # Calcular respuestas - CORREGIDO
    y_imp = np.array([y_impulso(ti) for ti in t])
    y_esc = np.array([y_escalon(ti) for ti in t])
    
    # Calcular estados para visualización - CORREGIDO
    x1_imp = np.array([x_impulso(ti)[0] for ti in t])
    x2_imp = np.array([x_impulso(ti)[1] for ti in t])
    x3_imp = np.array([x_impulso(ti)[2] for ti in t])
    
    x1_esc = np.array([x_escalon(ti)[0] for ti in t])
    x2_esc = np.array([x_escalon(ti)[1] for ti in t])
    x3_esc = np.array([x_escalon(ti)[2] for ti in t])
    
    # CREAR GRÁFICAS
    fig = plt.figure(figsize=(16, 12))
    
    # 1. Respuesta al impulso - Estados y salida
    ax1 = plt.subplot(3, 3, 1)
    ax1.plot(t, x1_imp, 'b-', linewidth=2, label='x₁(t)')
    ax1.plot(t, x2_imp, 'r-', linewidth=2, label='x₂(t)')
    ax1.plot(t, x3_imp, 'g-', linewidth=2, label='x₃(t)')
    ax1.set_title('Estados - Respuesta al Impulso')
    ax1.set_xlabel('Tiempo (s)')
    ax1.set_ylabel('Estados x(t)')
    ax1.grid(True, alpha=0.3)
    ax1.legend()
    
    # 2. Respuesta al escalón - Estados
    ax2 = plt.subplot(3, 3, 2)
    ax2.plot(t, x1_esc, 'b-', linewidth=2, label='x₁(t)')
    ax2.plot(t, x2_esc, 'r-', linewidth=2, label='x₂(t)')
    ax2.plot(t, x3_esc, 'g-', linewidth=2, label='x₃(t)')
    ax2.set_title('Estados - Respuesta al Escalón')
    ax2.set_xlabel('Tiempo (s)')
    ax2.set_ylabel('Estados x(t)')
    ax2.grid(True, alpha=0.3)
    ax2.legend()
    
    # 3. Salida y(t) - Comparación
    ax3 = plt.subplot(3, 3, 3)
    ax3.plot(t, y_imp, 'b-', linewidth=2, label='Impulso')
    ax3.plot(t, y_esc, 'r-', linewidth=2, label='Escalón')
    ax3.axhline(y=2, color='gray', linestyle='--', alpha=0.7, label='y(∞) = 2')
    ax3.set_title('Salida y(t) - Comparación')
    ax3.set_xlabel('Tiempo (s)')
    ax3.set_ylabel('y(t)')
    ax3.grid(True, alpha=0.3)
    ax3.legend()
    
    # 4. Salida y(t) - Impulso
    ax4 = plt.subplot(3, 3, 4)
    ax4.plot(t, y_imp, 'b-', linewidth=2)
    ax4.set_title('Salida y(t) - Respuesta al Impulso')
    ax4.set_xlabel('Tiempo (s)')
    ax4.set_ylabel('y(t)')
    ax4.grid(True, alpha=0.3)
    # Encontrar el máximo para posicionar el texto
    max_idx = np.argmax(y_imp)
    ax4.text(t[max_idx] + 0.5, y_imp[max_idx] * 0.8, 'Decae a cero', 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="blue", alpha=0.1))
    
    # 5. Salida y(t) - Escalón
    ax5 = plt.subplot(3, 3, 5)
    ax5.plot(t, y_esc, 'r-', linewidth=2)
    ax5.axhline(y=2, color='gray', linestyle='--', alpha=0.7, label='Valor final = 2')
    ax5.set_title('Salida y(t) - Respuesta al Escalón')
    ax5.set_xlabel('Tiempo (s)')
    ax5.set_ylabel('y(t)')
    ax5.grid(True, alpha=0.3)
    ax5.legend()
    ax5.text(3, 1.5, 'Converge a 2', 
             bbox=dict(boxstyle="round,pad=0.3", facecolor="red", alpha=0.1))
    
    # 6. Diagrama de polos
    ax6 = plt.subplot(3, 3, 6)
    # Polos en s = -1 (doble) y s = -2
    polos_x = [-1, -1, -2]
    polos_y = [0, 0, 0]
    ax6.scatter(polos_x, polos_y, color='red', s=100, marker='x', label='Polos', zorder=5)
    
    # Ejes y región estable
    ax6.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    ax6.axvline(x=0, color='k', linestyle='-', alpha=0.3, linewidth=2)
    
    # Sombrear región estable (semiplano izquierdo)
    ax6.axvspan(-3, 0, alpha=0.1, color='green', label='Región estable')
    
    ax6.set_xlabel('Parte Real (σ)')
    ax6.set_ylabel('Parte Imaginaria (jω)')
    ax6.set_title('Diagrama de Polos en Plano S')
    ax6.grid(True, alpha=0.3)
    ax6.set_xlim([-3, 1])
    ax6.set_ylim([-1.5, 1.5])
    ax6.legend()
    
    # 7. Matriz de transición Φ(t)
    ax7 = plt.subplot(3, 3, 7)
    t_phi = np.linspace(0, 4, 200)
    phi11 = np.exp(-t_phi)
    phi12 = t_phi * np.exp(-t_phi)
    phi22 = np.exp(-t_phi)
    phi33 = np.exp(-2*t_phi)
    
    ax7.plot(t_phi, phi11, 'b-', linewidth=2, label='Φ₁₁(t) = e⁻ᵗ')
    ax7.plot(t_phi, phi12, 'r-', linewidth=2, label='Φ₁₂(t) = te⁻ᵗ')
    ax7.plot(t_phi, phi22, 'g-', linewidth=2, label='Φ₂₂(t) = e⁻ᵗ')
    ax7.plot(t_phi, phi33, 'm-', linewidth=2, label='Φ₃₃(t) = e⁻²ᵗ')
    ax7.set_title('Elementos de Φ(t)')
    ax7.set_xlabel('Tiempo (s)')
    ax7.set_ylabel('Φ(t)')
    ax7.grid(True, alpha=0.3)
    ax7.legend()
    
    # 8. Comparación de convergencia
    ax8 = plt.subplot(3, 3, 8)
    ax8.semilogy(t, np.abs(y_imp), 'b-', linewidth=2, label='|y_imp(t)|')
    ax8.semilogy(t, np.abs(y_esc - 2), 'r-', linewidth=2, label='|y_esc(t) - 2|')
    ax8.set_title('Convergencia (Escala Log)')
    ax8.set_xlabel('Tiempo (s)')
    ax8.set_ylabel('Magnitud (log)')
    ax8.grid(True, alpha=0.3)
    ax8.legend()
    
    # 9. Valores finales de estados
    ax9 = plt.subplot(3, 3, 9)
    estados_finales = [2, 1, 0.5]  # x1(∞), x2(∞), x3(∞)
    labels = ['x₁(∞) = 2', 'x₂(∞) = 1', 'x₃(∞) = 0.5']
    colors = ['blue', 'red', 'green']
    bars = ax9.bar(labels, estados_finales, color=colors, alpha=0.7)
    ax9.set_title('Valores Finales de Estados\n(Respuesta al Escalón)')
    ax9.set_ylabel('Valor')
    
    # Añadir valores en las barras
    for bar, valor in zip(bars, estados_finales):
        height = bar.get_height()
        ax9.text(bar.get_x() + bar.get_width()/2., height,
                f'{valor}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    # ANÁLISIS NUMÉRICO
    print("\n" + "="*50)
    print("ANÁLISIS DEL SISTEMA (i)")
    print("="*50)
    
    print(f"\nPOLOS DEL SISTEMA:")
    print(f"• s = -1 (doble) - Polo múltiple")
    print(f"• s = -2 - Polo simple")
    
    print(f"\nESTABILIDAD:")
    print("✓ SISTEMA ESTABLE - Todos los polos en semiplano izquierdo")
    
    print(f"\nRESPUESTA AL IMPULSO:")
    print(f"• y(0) = {y_imp[0]:.3f}")
    print(f"• y(2) = {y_imp[250]:.3f}")
    print(f"• y(5) = {y_imp[625]:.3f}")
    print(f"• y(8) = {y_imp[-1]:.6f} → Decae a CERO")
    
    print(f"\nRESPUESTA AL ESCALÓN:")
    print(f"• y(0) = {y_esc[0]:.3f}")
    print(f"• y(2) = {y_esc[250]:.3f}")
    print(f"• y(5) = {y_esc[625]:.3f}")
    print(f"• y(8) = {y_esc[-1]:.3f} → Converge a 2")
    
    print(f"\nCONSTANTES DE TIEMPO:")
    print("• τ = 1 s (polo en -1)")
    print("• τ = 0.5 s (polo en -2)")
    
    print(f"\nOBSERVACIONES:")
    print("• Término te⁻ᵗ aparece por polo múltiple en s = -1")
    print("• Respuesta sobreamortiguada (polos reales)")
    print("• Sistema de fase mínima")
    print("• Buena rechazo a perturbaciones")
    
    # Verificación adicional
    print(f"\nVERIFICACIÓN DIMENSIONAL:")
    print(f"• Forma de x_impulso(0): {x_impulso(0).shape}")
    print(f"• Forma de y_impulso(0): {np.array([y_impulso(0)]).shape}")

# Ejecutar el análisis completo CORREGIDO
if __name__ == "__main__":
    sistema_tercer_orden_completo()
```
